```python
def categorize_col(df, col, labels):

    edges = [df[col].describe()["min"],
             df[col].describe()["25%"],
             df[col].describe()["50%"],
             df[col].describe()["75%"],
             df[col].describe()["max"]
             ]
    
    df[col]= pd.cut(df[col],edges, labels = labels)
```

```python
labels = ['not_popular','below_avegage','average','popular']
categorize_col(df,'Vote_Average',labels)
```
